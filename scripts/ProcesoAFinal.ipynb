```python
#Importamos librerias para la descarga

import datetime 
import os as os
import requests as rq
from lxml import html 
import time
import requests


os.chdir("C:\\Users\ortca\Analysic_Nabla\Proyectos\INFOSEL\ProcesoA\FondosPDF_root\BBVA")
#urls de las 4 pags de bbva [Fondos renta variable,Fondos de deuda,Fondos de estrategias,Fondos multiestrategia ]
urls = ['https://www.bbva.mx/personas/productos/inversion/normativa-de-fondos/fondos-renta-variable.html','https://www.bbva.mx/personas/productos/inversion/normativa-de-fondos/fondos-de-deuda.html',
      'https://www.bbva.mx/personas/productos/inversion/normativa-de-fondos/fondos-de-estrategias.html','https://www.bbva.mx/personas/productos/inversion/normativa-de-fondos/fondos-de-multiestrategias.html']

#Creamos la carpeta donde se guardarán los DICIS  DICISBBVA_Año_Mes_dia
hoy=str(datetime.date.today())
direcactual=os.getcwd()
os.mkdir(hoy)
os.chdir("C:\\Users\ortca\Analysic_Nabla\Proyectos\INFOSEL\ProcesoA\FondosPDF_root\BBVA\\"+hoy)
os.mkdir("DicisBBVA")
os.mkdir("ProspectosBBVA")



#Ciclo que entra a cada url y extrae la información 
for i in urls:
    url=i
    encabezado = {
    'user-agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/71.0.3578.80 Chrome/71.0.3578.80 Safari/537.36'
        }
    respuesta = rq.get(url,headers=encabezado)
    parser = html.fromstring(respuesta.text)
    linksdici =parser.xpath('//div[@id="par-table"]//tr//td[position()=3]//div[@class="table--value"]//a/@href')
    linksprospec =parser.xpath('//div[@id="par-table"]//tr//td[position()=2]//div[@class="table--value"]//a/@href')
    fondos =parser.xpath('//div[@id="par-table"]//tr//td[position()=1]//div[@class="table--value"]/text()')
    fontot= [x for x in list(map(lambda x: x.replace("\n","").strip(),fondos)) if x!='']
    try:
        os.chdir(direcactual+"\\"+hoy+"\\DicisBBVA")
        j = 0
        for k in linksdici:
            urldi=str(k)
            page = requests.get(urldi)
            time.sleep(3)
            with open(fontot[j]+".pdf", "wb") as f:
                f.write(page.content)
            j+=1
            
        os.chdir(direcactual+"\\"+hoy+"\\ProspectosBBVA")
        l = 0
        for m in linksprospec:
            urlpro=str(m)
            page = requests.get(urlpro)
            time.sleep(3)
            with open(fontot[l]+".pdf", "wb") as f:
                f.write(page.content)
            l+=1
        print("url -- cargada")
    
    except Exception as e:
        os.chdir(direcactual)
        print(f"error fondos BBVA: fondo {fontot[j]}")
        print(e)
os.chdir(direcactual)
```
